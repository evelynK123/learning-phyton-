```python
#library bla bla bla 
#sql injection untuk bambil data base di API (cybersecurity)
import pandas as pd 
from bs4 import BeautifulSoup 
import requests
```

```python
url = "https://quotes.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
print(soup)
```

```python
#2. ambil semua kutipan dan juga penulisnya 

'''
bantuan tag hmtl 
<div class="quote" itemscope="" itemtype="http://schema.org/CreativeWork">
<span class="text" itemprop="text">“The person, be it gentleman or lady, who has not pleasure in a good novel, must be intolerably stupid.”</span>
<span>by <small class="author" itemprop="author">Jane Austen</small>
<a href="/author/Jane-Austen">(about)</a>
kutipan | penulis 
'''

quotes = [] # variable list kosong 


for q in soup.find_all("div", class_="quote"):
    text = q.find("span", class_="text").text
    penulis = q.find("small", class_="author").text
    #print(f"Nama: {penulis}, \nTulisan: {text}")
    quotes.append({'text': text, 'author': penulis})
```

```python
#quotes

#disimpan dalam data frame data ke csv
# bisa dalam bentuk json (bisa pake pandas), excel, html, markdown
#world cloud

df = pd.DataFrame(quotes)
df.to_csv("kutipan.csv", index=False) #menhapus index
print("Data berhasil disimpan ke kutipan.csv")

```
