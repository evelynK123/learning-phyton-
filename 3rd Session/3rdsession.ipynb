```python
#selisih, gabungan, irisan 
nama_siswa_pertama = {'Kevin', 'Ario', 'Evelyn', 'Jacob', 'Brandon', 'Micheal', 'Charlie'}
nama_siswa_kedua = {'Kevin', 'Micheal', 'Jacob', 'Brandon', 'Chaplin', "Trump"}

#print dua duanya tapi masih dalam satu set. or atau |
print(nama_siswa_pertama | nama_siswa_kedua)

#selisih, yang gk ada di set pertama (0)
print(nama_siswa_pertama - nama_siswa_kedua)

# irirsan, yang ada di keduanya
print(nama_siswa_pertama & nama_siswa_kedua)
```
