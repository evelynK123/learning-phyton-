```python
''' 
ERROR HANDLING!
bug itu kesalahan 
klik share malah menampilkan file 
kalau error 
input meminta angka tapi kita memasuka huruf contoh 
string tambah string atau string tambah integer error
'''

import tkinter as tk 
from tkinter import ttk, messagebox #box notifikasi 

def hitung():
    try:
        a = float(entry1.get())
        b = float(entry2.get()) # float data desimal
        hasil = a / b
        label_hasil.config(text=f"Hasil: {hasil}")
    except ValueError:
        messagebox.showerror("input salah!", "harap masukan nilai yang valid!")
    except ZeroDivisionError:
        messagebox.showerror("error matematika", "tidak bisa membagi dengan nol")

window = tk.Tk()
window.title("error handling demo")
window.geometry("300x150")

entry1 = tk.Entry(window, width=20)
entry1.pack(pady=5)
entry2 = tk.Entry(window, width=20)   
entry2.pack(pady=5)
        
button = tk.Button(window, text='Hitung', command=hitung)
button.pack(pady=5)

label_hasil = tk.Label(window,text="Masukkan dua angka lalu klik hitung")
label_hasil.pack(pady=5)

window.mainloop()
```
